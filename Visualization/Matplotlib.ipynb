```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10 ,1000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y,'--b')
plt.grid()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sin curve')
```

```python
x = np.random.rand(50)
y =np.random.rand(50)
```

```python
colours = np.random.rand(50)
size =1000*np.random.rand(50)
plt.scatter(x,y,c=colours,s=size)
plt.title('random variable')
plt.xlabel('X-axis')
```
