### SMOTES(Synthetic Minority Oversampling Technique)

SMOTE (Synthetic Minority over -Sampling Technique) is technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the manjority class. SMOTE involved generating instances of the minority class by interpolating between existing instances


```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[.90])
```

```python
import pandas as pd
```

```python
df1= pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## transform the datas
oversample =SMOTE()
x,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```
